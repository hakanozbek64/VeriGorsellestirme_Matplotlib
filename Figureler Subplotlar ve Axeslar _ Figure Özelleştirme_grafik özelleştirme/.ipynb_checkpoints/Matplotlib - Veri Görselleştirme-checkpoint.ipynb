```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6)
y = np.arange(1,11,2)

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.2,0.3])

axes1.plot(y,x)

axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outher Graph")

axes2.plot(x,y)

axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")



plt.show()
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1)
                  
axes[0].plot(x,y)
axes[0].set_title("First Axes")
                  
axes[1].set_title("Second  Axes")                  
axes[1].plot(x,y**3)
    
plt.tight_layout()
```

```python
fig =plt.figure(figsize = (8,6))

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = "purple")
```

```python
fig,axes =plt.subplots(nrows =2,ncols =1,figsize =(8,6))

axes[0].plot(x,y,color = "red")
axes[0].set_title("First Axes")
axes[1].plot(x,x**0.5,color = "blue")
axes[1].set_title("Second Axes")
plt.tight_layout()
```

```python
fig = plt.figure(figsize =(6,4))

axes =fig.add_axes([0,0,1,1])

axes.plot(x,y**.5,color = "red",label = "x karekök")
axes.plot(x,x**2,color = "yellow",label ="x kare")
axes.plot(x,x**3,color = "#eabcac",label ="x küp")
axes.legend()
fig.savefig("Figure2.png")
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color = "blue",linewidth = 3,marker = "o",markersize = "15",markerfacecolor = "black"
         ,markeredgecolor ="yellow",markeredgewidth = 5)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x**2,color = "red",linewidth = 3,marker = "o",markersize = "10",markerfacecolor = "black"
         ,markeredgecolor ="yellow",markeredgewidth = 5)

axes.set_xlim(0,10)
axes.set_ylim(0,50)
```
